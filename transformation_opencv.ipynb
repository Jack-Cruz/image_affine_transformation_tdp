```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

## Scaling

```python
# Resizing using fx, fy parameters

img = cv.imread('left.png')

#res = cv.resize(img, None, fx=1, fy=2, interpolation=cv.INTER_CUBIC)
res = cv.transpose(img)

plt.imshow(res)
```

```python
# Resizing using specified size

img = cv.imread('left.png')

height, width = img.shape[:2]
res = cv.resize(img, (1*width, 2*height), interpolation=cv.INTER_CUBIC)

plt.imshow(res)
```

## Translation

```python
img = cv.imread('left.png')
rows, cols = img.shape[:2]

M = np.float32([[1.0, 0, -100],
                [0,   1,  50]])
dst = cv.warpAffine(img, M, (cols, rows))

plt.imshow(dst)
```

## Rotation

```python
img = cv.imread('left.png')
rows, cols = img.shape[:2]

# cols-1 and rows-1 are the coordinate limits.
M = cv.getRotationMatrix2D(((cols-1)/2.0, (rows-1)/2.0), 90, 1)
dts = cv.warpAffine(img, M, (cols, rows))

plt.imshow(dts)
```

## Affine Transformation

```python
img = cv.imread('left.png')
rows, cols, ch = img.shape

print('rows:', rows)
print('cols:', cols)
print('rows-mid:', (rows-1)/2.0)
print('cols-mid:', (cols-1)/2.0)


pts1 = np.float32([[0, 0], [1125, 0], [0, 369]])
pts2 = np.float32([[0, 100], [1000, 0], [0, 369]])

M = cv.getAffineTransform(pts1, pts2)
dst = cv.warpAffine(img, M, (cols, rows))

plt.subplot(121)
plt.imshow(img)
plt.title('Input')

plt.subplot(122)
plt.imshow(dst)
plt.title('output')
plt.show()
```

## Perspective transformation

```python
img = cv.imread('left.png')
rows, cols, ch = img.shape

pts1 = np.float32([[56, 56], [368, 52], [28, 387], [389, 390]])
pts2 = np.float32([[0, 0], [300, 0], [0, 300], [300, 300]])

M = cv.getPerspectiveTransform(pts1, pts2)
dst = cv.warpPerspective(img, M, (300,300))

plt.subplot(121), plt.imshow(img), plt.title('Input')
plt.subplot(122), plt.imshow(dst), plt.title('Output')
plt.show()
```
