## Accessing and Modifying pixel values

```python
# Load color image

img = cv.imread('./left.png')
px = img[100, 100]
print('RGB:', px)   # It returns an array of Blue, Green, Red values or just intensity in Grayscale image

# Accessing only blue value

blue = img[100, 100, 0]
print('B:',blue)

# Modify the pixel values

img[100, 100] = [255, 255, 255]
print('pixel changed:', img[100, 100])
```

```python
# Better accessing RED value
img.item(10, 10, 2)

# Modifying RED value
img.itemset((10, 10, 2), 100)
img.item(10, 10, 2)
```

# OpenCV + Matplotlib

```python
# Printing an RGB image 

# Import required module
import cv2
import matplotlib.pyplot as plt

# Read Image
image = cv2.imread('left.png')

# Call imshow() using plt object
plt.imshow(image)

# Display that image
plt.show()
```

```python
# Printing a Gray scaled image
img1 = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(img1, cmap='gray')
plt.show()
```
